### Splitting the dataset by numerical and categorical columns using loop

```python
# Create empty lists to store numerical and categorical column names
numerical_cols = []
categorical_cols = []

# Iterate through the columns and check their data types
for column in df.columns:
    if df[column].dtype in [int, float]:  # Check if the data type is numerical
        numerical_cols.append(column)
    else:
        categorical_cols.append(column)

# Now, you have two lists: numerical_cols and categorical_cols
print("Numerical Columns:", numerical_cols)
print("Categorical Columns:", categorical_cols)
```

### Checking for null values

```python
df.isnull().sum()
```

```python
from sklearn.impute import SimpleImputer

# Create a SimpleImputer for numerical columns
numerical_imputer = SimpleImputer(strategy='mean')

# Select the numerical columns you want to impute
numerical_columns = ['Age', '90s', 'Gls', 'Ast', 'G-PK', 'PK', 'PKatt', 'CrdY', 'CrdR', 'Gls.1', 'Ast.1', 'G+A', 'G-PK.1', 'G+A-PK', 'xG', 'npxG', 'xA', 'npxG+xA', 'xG.1', 'xA.1', 'xG+xA', 'npxG.1', 'npxG+xA.1']

# Impute missing values in the selected numerical columns
df[numerical_columns] = numerical_imputer.fit_transform(df[numerical_columns])
```

```python
df
```

```python
from sklearn.impute import SimpleImputer

# Create a SimpleImputer for categorical columns with the 'most_frequent' strategy
categorical_imputer = SimpleImputer(strategy='most_frequent')

# Select the categorical columns you want to impute
categorical_columns = ['Player', 'Team', 'Nation', 'Pos', 'MP', 'Starts', 'Min']

# Impute missing values in the selected categorical columns
df[categorical_columns] = categorical_imputer.fit_transform(df[categorical_columns])
```

### Expected Goals Analysis

####  Is there a correlation between actual goals (Gls) and expected goals (xG)?

```python
# Calculate the correlation coefficient between Gls and xG
correlation_coefficient = df['Gls'].corr(df['xG'])

# Display the correlation coefficient
print(f"Correlation coefficient between Gls and xG: {correlation_coefficient:.2f}")
```

##### what this code is pointing out is that;
- The correlation coefficient can range from -1 to 1.
- A positive value closer to 1 indicates a strong positive correlation (as actual goals increase, expected goals also increase).
- A negative value closer to -1 indicates a strong negative correlation (as actual goals increase, expected goals decrease).
- A value close to 0 indicates a weak or no linear correlation.

- The result shows that there is a correlation between actual goals (Gls) and expected goals (xG) which is 0.93.

```python
# Ensure 'Age' column is in numerical format
df['Age'] = df['Age'].astype(float)

# Calculate Pearson correlation coefficients
age_goals_corr = df['Age'].corr(df['Gls'])
age_assists_corr = df['Age'].corr(df['Ast'])

# Create scatter plots
plt.figure(figsize=(12, 6))

# Scatter plot for age vs. goals
plt.subplot(1, 2, 1)
sns.scatterplot(x='Age', y='Gls', data=df)
plt.title(f'Age vs. Goals (Correlation: {age_goals_corr:.2f})')

# Scatter plot for age vs. assists
plt.subplot(1, 2, 2)
sns.scatterplot(x='Age', y='Ast', data=df)
plt.title(f'Age vs. Assists (Correlation: {age_assists_corr:.2f})')

plt.tight_layout()
plt.show()
```

#### Analyze the relationship between age and disciplinary actions (yellow and red cards)

```python
# Create age bins
age_bins = [18, 23, 28, 33, 38, 43, 48]
age_labels = ['18-23', '24-28', '29-33', '34-38', '39-43', '44-48']

# Add age group labels to the DataFrame
df['Age Group'] = pd.cut(df['Age'], bins=age_bins, labels=age_labels, right=False)

# Calculate the total number of yellow and red cards for each age group
age_disciplinary_actions = df.groupby('Age Group')[['CrdY', 'CrdR']].sum().reset_index()

# Sort the age groups by age in ascending order
age_disciplinary_actions_sorted = age_disciplinary_actions.sort_values(by='Age Group', ascending=True)

# Create a bar chart to visualize the relationship between age and disciplinary actions
plt.figure(figsize=(12, 8))
x = age_disciplinary_actions_sorted['Age Group']
plt.bar(x, age_disciplinary_actions_sorted['CrdY'], color='gold', label='Yellow Cards')
plt.bar(x, age_disciplinary_actions_sorted['CrdR'], bottom=age_disciplinary_actions_sorted['CrdY'], color='red', label='Red Cards')

plt.xlabel('Age Group')
plt.ylabel('Total Cards')
plt.title('Relationship Between Age and Disciplinary Actions')
plt.legend()

plt.show()
```
